# Gender and User type relation

```python
total_men=df2.query(' member_gender == "Male" ')
total_women=df2.query('member_gender == "Female"' )
total_other=df2.query('member_gender == "Other"')
total_men_Subscriber=total_men.query('user_type == "Subscriber"').count()[0]
total_men_customer=total_men.count()[0]-total_men_Subscriber
total_women_Subscriber=total_women.query('user_type == "Subscriber"').count()[0]
total_women_customer=total_women.count()[0]-total_women_Subscriber
```

```python
total_women_customer
```

```python
#Men 
plt.figure(figsize=[ 11.69, 8.27])
plt.subplot(3, 1, 1)  
n_points = total_men.shape[0]
max_count = total_men['member_gender'].value_counts().max()
max_prop = max_count / n_points
tick_props = np.arange(0, max_prop+.1, 0.1)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]
sb.countplot(data= total_men,x='member_gender',hue='user_type');
plt.yticks(tick_props * n_points, tick_names);
plt.ylabel('Proportion (%)');
plt.xlabel('Male User');
plt.title('Male and User type relation');
#Woman
plt.subplot(3, 1, 2) 
n_points = total_women.shape[0]
max_count = total_women['member_gender'].value_counts().max()
max_prop = max_count / n_points
tick_props = np.arange(0, max_prop+.1, 0.1)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]
sb.countplot(data= total_women,x='member_gender',hue='user_type');
plt.yticks(tick_props * n_points, tick_names);
plt.ylabel('Proportion (%)');
plt.xlabel('Female User');
plt.title('Female and User type relation');
#Other 
plt.subplot(3, 1, 3) 
n_points = total_other.shape[0]
max_count = total_other['member_gender'].value_counts().max()
max_prop = max_count / n_points
tick_props = np.arange(0, max_prop+.1, 0.1)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]
sb.countplot(data= total_other,x='member_gender',hue='user_type');
plt.yticks(tick_props * n_points, tick_names);
plt.ylabel('Proportion (%)');
plt.xlabel('Other User');
plt.title('Other and User type relation');
```
